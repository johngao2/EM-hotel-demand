```python
import pandas as pd
import numpy as np
import os
from datetime import timedelta, date

'''This script produces a file which shows transactions. Each row is a transaction
with arrival, depart, and room type.'''

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

# load raw hotel data
df = pd.read_csv('data/2017_booking_data.csv', parse_dates=['ARRIVAL', 'DEPART', 'DATE'])
df = df.sort_values(['DATE', 'RESNO'])
```

```python
# find booking begin and end dates
df_grouped_book = df.groupby(by='DATE').count()
book_start = df_grouped_book.index.min()
book_end = df_grouped_book.index.max()

# find season begin and end dates
df_grouped_arr = df.groupby(by='ARRIVAL').count()
ssn_start = df_grouped_arr.index.min()
ssn_end = df_grouped_arr.index.max()

# generate list of season dates
ssn = pd.date_range(ssn_start, ssn_end)

# generate room type list
df_grouped_types = df.groupby(by='RM TYPE').count()
rm_types = df_grouped_types.index.tolist()

# create blank df
df2 = pd.DataFrame(index = pd.date_range(book_start, book_end),
                   columns = pd.MultiIndex.from_product([rm_types, ssn], 
                                                        names=['TYPE', 'STAY DATE']))
df2 = df2.fillna(0)
```
